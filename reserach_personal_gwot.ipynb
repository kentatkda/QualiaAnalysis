```python
import numpy as np
from research_nonal_al import run_nonal_al_all
from research_nonal import run_nonal_all
non_al_folder_path = './qualia_rawdata/non_alcohol/'
al_folder_path = './qualia_rawdata/alcohol/'
qualia_color = {
    'empatic_pain': 'orange',
    'guilt': 'black',
    'fear': 'violet',
    'anger': 'red',
    'envy': 'purple',
    'sadness': 'blue',
    'surprise': 'gray',
    'joy': 'cyan',
    'amusement': 'brown',
    'romance': 'pink',
    'aesthetic_appreciation': 'yellow',
    'awe': 'green',
}
```

## non_al vs al の被験者ごとに見てみる

```python
optimal_P1, optimal_Q1, all_original_embeddings1, all_embeddings1, distances_df1 = run_nonal_al_all(
    non_al_folder_path=non_al_folder_path,
    al_folder_path=al_folder_path,
    iter_num=50,
    qualia_color=qualia_color
)
```

```python
list(dict(distances_df1.mean(axis=1)).values())
```

```python
import matplotlib.pyplot as plt
distances = list(dict(distances_df1.mean(axis=1)).values())
emotions = list(qualia_color.keys())
plt.figure(figsize=(7,4))
plt.bar(emotions, distances, color='skyblue')
plt.xlabel('Emotions')
plt.ylabel('Distance')
plt.title('Average Distance for Each Emotion')
plt.xticks(emotions, rotation=45, ha='right')
plt.grid(axis='y')
```
